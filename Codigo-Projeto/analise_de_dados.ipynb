```python
# passo 1: importar base de dados e verificar quais dados e suas quantidades.

import pandas as pd
pd.set_option('display.max_rows', 700)
tabela = pd.read_csv('codevasf_202312281536.csv')

# passo 2: remoção da coluna estação pois não e um dado a seguir como padrão
tabela = tabela.drop(columns=['estacao'])

# passo 3: verificado usuarios e quais produtos foram acessados 
dt = tabela.pivot_table(columns=['usuario', 'produto'], aggfunc='size')

# Passo 4: ordenando usuarios e seus respectivos produtos acessados para excluir duplicados 
tabela = tabela.sort_values(by='usuario')


```

```python
# Passo 4: remover dados de usuarios que acessou o mesmo produto mais de uma vez
tabela = tabela.drop_duplicates(['usuario', 'produto'])


# passo 5: remover dados de usuários duplicados e manter o ultimo acesso ao produto mais atual
dados = tabela.drop_duplicates(subset=['usuario'], keep='first')

# passo 6: verificar se todos tados duplicados de usuarios foram removidos 
dt = dados.pivot_table(columns=['usuario', 'produto'], aggfunc='size')
display(dt)
# passo 7: verificar se quantidade de registros diminuiu consideravelmente 
display(dt.info())
display(dt.groupby('produto').size().sort_values(ascending=True))
```
